```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from pathlib import Path

print("=" * 80)
print("HABITS DATASET - STRICT CLEANING & STANDARDIZATION FOR ML")
print("=" * 80)

# Load the cleaned habits dataset
df = pd.read_csv(r'c:\vscode\CVprojects\kenexaiHackathon\cleaned datasets\cleaned_habits_dataset.csv')

print("\nINITIAL DATASET")
print("-" * 80)
print(f"Shape: {df.shape}")
print(f"Columns ({len(df.columns)}): {df.columns.tolist()}")
print(f"\nFirst row:")
print(df.iloc[0])
```

```python
# Step 1: Remove duplicate rows
print("\n1. REMOVE DUPLICATE ROWS")
print("-" * 80)

duplicates_before = len(df)
df = df.drop_duplicates().reset_index(drop=True)
duplicates_removed = duplicates_before - len(df)

print(f"Rows before: {duplicates_before:,}")
print(f"Duplicates removed: {duplicates_removed}")
print(f"Rows after: {len(df):,}")
```

```python
# Step 2: Standardize column names
print("\n2. STANDARDIZE COLUMN NAMES")
print("-" * 80)

print("Converting column names to lowercase and replacing spaces with underscores...")
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('[^a-z0-9_]', '', regex=True)

print(f"\nStandardized columns ({len(df.columns)}):")
print(df.columns.tolist())
```

```python
# Step 3: Standardize categorical values
print("\n3. STANDARDIZE CATEGORICAL VALUES")
print("-" * 80)

# Standardize gender
if 'gender' in df.columns:
    df['gender'] = df['gender'].astype(str).str.lower().str.strip()
    df['gender'] = df['gender'].map(lambda x: 'male' if 'male' in x or 'm' == x else ('female' if 'female' in x or 'f' == x else 'other'))
    print(f"✓ gender: {df['gender'].unique().tolist()}")

# Standardize diet_quality
if 'diet_quality' in df.columns:
    df['diet_quality'] = df['diet_quality'].astype(str).str.lower().str.strip()
    diet_map = {'poor': 'poor', 'fair': 'fair', 'good': 'good', 'bad': 'poor', 'avg': 'fair', 'average': 'fair'}
    df['diet_quality'] = df['diet_quality'].map(lambda x: diet_map.get(x, x) if x in diet_map else x)
    print(f"✓ diet_quality: {df['diet_quality'].unique()[:5]}")

# Standardize internet_quality
if 'internet_quality' in df.columns:
    df['internet_quality'] = df['internet_quality'].astype(str).str.lower().str.strip()
    int_map = {'low': 'low', 'medium': 'medium', 'high': 'high', 'poor': 'low', 'good': 'high', 'fair': 'medium'}
    df['internet_quality'] = df['internet_quality'].map(lambda x: int_map.get(x, x) if x in int_map else x)
    print(f"✓ internet_quality: {df['internet_quality'].unique().tolist()}")

# Standardize study_environment
if 'study_environment' in df.columns:
    df['study_environment'] = df['study_environment'].astype(str).str.lower().str.strip()
    print(f"✓ study_environment: {df['study_environment'].unique()[:5]}")

# Standardize parental_education_level
if 'parental_education_level' in df.columns:
    df['parental_education_level'] = df['parental_education_level'].astype(str).str.lower().str.strip()
    print(f"✓ parental_education_level: {df['parental_education_level'].unique()[:5]}")

# Standardize family_income_range
if 'family_income_range' in df.columns:
    df['family_income_range'] = df['family_income_range'].astype(str).str.lower().str.strip()
    income_map = {'low': 'low', 'medium': 'medium', 'high': 'high', 'middle': 'medium', 'upper': 'high', 'lower': 'low'}
    df['family_income_range'] = df['family_income_range'].map(lambda x: income_map.get(x, x) if x in income_map else x)
    print(f"✓ family_income_range: {df['family_income_range'].unique().tolist()}")

print(f"\nCategorical standardization complete")
```

```python
# Step 4: Convert binary columns to integers
print("\n4. CONVERT BINARY COLUMNS TO INTEGERS")
print("-" * 80)

binary_cols = ['part_time_job', 'access_to_tutoring', 'extracurricular_participation', 'dropout_risk']

for col in binary_cols:
    if col in df.columns:
        # Convert to int (0/1)
        df[col] = df[col].astype(int)
        print(f"✓ {col}: {sorted(df[col].unique())}")

print(f"\nBinary columns converted to integers")
```

```python
# Step 5: Ensure numeric columns are float or integer
print("\n5. ENSURE NUMERIC COLUMNS ARE PROPER TYPE")
print("-" * 80)

numeric_cols = ['age', 'semester', 'mental_health_rating', 'motivation_level', 
                'time_management_score', 'exam_anxiety_score', 'final_exam_marks', 'gpa']

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        print(f"✓ {col}: {df[col].dtype}")

# Also convert all other numeric columns
for col in df.columns:
    if col not in numeric_cols and col not in ['gender', 'major', 'diet_quality', 
                                                 'parental_education_level', 'internet_quality',
                                                 'learning_style', 'study_environment', 
                                                 'family_income_range', 'source_dataset']:
        if df[col].dtype == 'object':
            try:
                df[col] = pd.to_numeric(df[col], errors='ignore')
            except:
                pass

print(f"\nNumeric type conversion complete")
```

```python
# Step 6: Recalculate derived features
print("\n6. RECALCULATE DERIVED FEATURES")
print("-" * 80)

# total_screen_time = social_media_hours + netflix_hours
if 'social_media_hours' in df.columns and 'netflix_hours' in df.columns:
    df['total_screen_time'] = df['social_media_hours'] + df['netflix_hours']
    print(f"✓ total_screen_time = social_media_hours + netflix_hours")
    print(f"  Range: {df['total_screen_time'].min():.2f} - {df['total_screen_time'].max():.2f}")

# productivity_score = study_hours_per_day / screen_time
if 'study_hours_per_day' in df.columns and 'screen_time' in df.columns:
    df['productivity_score'] = df['study_hours_per_day'] / (df['screen_time'] + 0.1)
    print(f"✓ productivity_score = study_hours_per_day / screen_time")
    print(f"  Range: {df['productivity_score'].min():.4f} - {df['productivity_score'].max():.4f}")

# wellbeing_score = (sleep_hours + exercise_frequency + mental_health_rating) / 3
if all(col in df.columns for col in ['sleep_hours', 'exercise_frequency', 'mental_health_rating']):
    df['wellbeing_score'] = (df['sleep_hours'] + df['exercise_frequency'] + df['mental_health_rating']) / 3
    print(f"✓ wellbeing_score = (sleep_hours + exercise_frequency + mental_health_rating) / 3")
    print(f"  Range: {df['wellbeing_score'].min():.2f} - {df['wellbeing_score'].max():.2f}")

print(f"\nDerived features recalculated")
```

```python
# Step 7: Normalize all numeric features using MinMaxScaler
print("\n7. NORMALIZE NUMERIC FEATURES (MinMaxScaler)")
print("-" * 80)

# Get all numeric columns except integers that represent binary/categories
numeric_dtypes = df.select_dtypes(include=['float64', 'int64']).columns
# Exclude binary columns and integer category columns
exclude = ['part_time_job', 'access_to_tutoring', 'extracurricular_participation', 'dropout_risk', 'semester']
scale_cols = [col for col in numeric_dtypes if col not in exclude]

print(f"Normalizing {len(scale_cols)} numeric features...")

if scale_cols:
    scaler = MinMaxScaler()
    df[scale_cols] = scaler.fit_transform(df[scale_cols])
    print(f"✓ Features normalized to [0, 1] range")
    print(f"\nNormalized columns:")
    for col in scale_cols[:10]:
        print(f"  {col}: min={df[col].min():.4f}, max={df[col].max():.4f}")
    if len(scale_cols) > 10:
        print(f"  ... and {len(scale_cols) - 10} more columns")

print(f"\nNormalization complete")
```

```python
# Step 8: Check and fill missing values
print("\n8. CHECK AND FILL MISSING VALUES")
print("-" * 80)

# Check for missing values
missing_count = df.isnull().sum()
missing_pct = (missing_count / len(df)) * 100
missing_data = pd.DataFrame({
    'column': missing_count.index,
    'missing_count': missing_count.values,
    'missing_pct': missing_pct.values
})
missing_data = missing_data[missing_data['missing_count'] > 0].sort_values('missing_pct', ascending=False)

if len(missing_data) > 0:
    print("Missing values found:")
    print(missing_data.to_string(index=False))
    
    # Fill numeric columns with median
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for col in numeric_cols:
        if df[col].isnull().sum() > 0:
            median_val = df[col].median()
            df[col].fillna(median_val, inplace=True)
            print(f"  Filled {col} with median: {median_val:.4f}")
    
    # Fill categorical columns with mode
    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        if df[col].isnull().sum() > 0:
            mode_val = df[col].mode()[0] if not df[col].mode().empty else 'Unknown'
            df[col].fillna(mode_val, inplace=True)
            print(f"  Filled {col} with mode: '{mode_val}'")
else:
    print("✓ No missing values found")

print(f"\nFinal missing values: {df.isnull().sum().sum()}")
```

```python
# Step 9: Print final dataset info and summary
print("\n9. FINAL DATASET SUMMARY")
print("=" * 80)

print(f"\nDataset Shape: {df.shape}")
print(f"Records: {len(df):,}")
print(f"Features: {len(df.columns)}")

print(f"\n\nData Quality Metrics:")
print(f"  Missing values: {df.isnull().sum().sum()}")
print(f"  Duplicate rows: {df.duplicated().sum()}")
print(f"  Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

print(f"\n\nData Types Distribution:")
print(df.dtypes.value_counts())

print(f"\n\nAll Columns ({len(df.columns)}):")
for i, col in enumerate(df.columns, 1):
    dtype = df[col].dtype
    print(f"  {i:2d}. {col:40s} ({dtype})")

print(f"\n\nBasic Statistics:")
print(df.describe())
```

```python
# Step 10: Save the final cleaned dataset
print("\n10. SAVE FINAL DATASET")
print("=" * 80)

output_path = Path(r'c:\vscode\CVprojects\kenexaiHackathon\cleaned datasets\cleaned_habits_dataset_final.csv')
output_path.parent.mkdir(parents=True, exist_ok=True)

df.to_csv(output_path, index=False)

print(f"\n✓ Final dataset saved: {output_path}")
print(f"\nFile Details:")
print(f"  Path: {output_path}")
print(f"  Shape: {df.shape}")
print(f"  Size: {output_path.stat().st_size / 1024 / 1024:.2f} MB")

print(f"\n" + "=" * 80)
print("STRICT CLEANING & STANDARDIZATION COMPLETE - DATASET READY FOR ML")
print("=" * 80)
print(f"\n✓ All 10 steps completed successfully!")
print(f"✓ Dataset is now standardized and ready for machine learning")
```

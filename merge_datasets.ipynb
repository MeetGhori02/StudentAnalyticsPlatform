```python
import pandas as pd
import numpy as np
from pathlib import Path

print("=" * 80)
print("STUDENT SUCCESS DATASET - MERGE & DATA STITCHING PIPELINE")
print("=" * 80)

# Section 1: Load the Three Cleaned Datasets
print("\n1. LOADING CLEANED DATASETS FROM cleaned datasets/ FOLDER")
print("-" * 80)

# Load datasets
df_attitude = pd.read_csv(r'c:\vscode\CVprojects\kenexaiHackathon\cleaned datasets\cleaned_attitude_dataset.csv')
df_research = pd.read_csv(r'c:\vscode\CVprojects\kenexaiHackathon\cleaned datasets\research_student_cleaned.csv')
df_habits = pd.read_csv(r'c:\vscode\CVprojects\kenexaiHackathon\cleaned datasets\cleaned_habits_dataset.csv')

print(f"✓ Attitude Dataset: {df_attitude.shape}")
print(f"✓ Research (Engineering) Dataset: {df_research.shape}")
print(f"✓ Habits & Performance Dataset: {df_habits.shape}")
```

```python
# Section 2: Explore Dataset Structures
print("\n2. EXPLORE DATASET STRUCTURES")
print("-" * 80)

print(f"\nAttitude Dataset:")
print(f"  Shape: {df_attitude.shape}")
print(f"  Columns: {df_attitude.columns.tolist()}")

print(f"\n\nResearch Dataset:")
print(f"  Shape: {df_research.shape}")
print(f"  Columns: {df_research.columns.tolist()}")

print(f"\n\nHabits Dataset:")
print(f"  Shape: {df_habits.shape}")
print(f"  Columns (first 20): {df_habits.columns.tolist()[:20]}")
```

```python
# Section 3: Standardize Column Names Across Datasets
print("\n3. STANDARDIZE COLUMN NAMES ACROSS DATASETS")
print("-" * 80)

# Create copies to avoid modifying original datasets
attitude = df_attitude.copy()
research = df_research.copy()
habits = df_habits.copy()

# Map common features across datasets
# For Attitude dataset: rename marks columns
if 'marks' in attitude.columns:
    attitude = attitude.rename(columns={'marks': 'academic_average'})

# For Research dataset: rename marks columns to create academic_average
research['academic_average'] = (research['marks10th'] + research['marks12th']) / 2
research = research.rename(columns={'cgpa': 'gpa', 'branch': 'department'})

# For Habits dataset: rename columns for consistency
habits = habits.rename(columns={'major': 'department', 'exam_score': 'final_exam_marks', 
                                'previous_gpa': 'gpa'})

# Add missing 'gpa' column to attitude dataset if needed
if 'gpa' not in attitude.columns and 'marks' in attitude.columns:
    # Estimate GPA based on marks (rough conversion)
    attitude['gpa'] = attitude['marks'] / 20.0  # Assuming marks out of 100

# Ensure all datasets have these standard columns
standard_cols = ['gender', 'academic_average', 'gpa', 'department']
print(f"Standard columns to align: {standard_cols}")

print(f"\nAttitude Dataset columns after standardization:")
print([col for col in attitude.columns if col in standard_cols or 'academic' in col or 'gpa' in col])

print(f"\nResearch Dataset columns after standardization:")
print([col for col in research.columns if col in standard_cols or 'academic' in col or 'gpa' in col])

print(f"\nHabits Dataset columns after standardization:")
print([col for col in habits.columns if col in standard_cols or 'academic' in col or 'gpa' in col])
```

```python
# Section 4: Align Common Features
print("\n4. ALIGN COMMON FEATURES")
print("-" * 80)

# Check gender values across datasets
print(f"\nGender values in each dataset:")
print(f"  Attitude: {attitude['gender'].unique()[:5]}")
print(f"  Research: {research['gender'].unique()[:5]}")
print(f"  Habits: {habits['gender'].unique()[:5] if 'gender' in habits.columns else 'N/A'}")

# Check academic marks/average
print(f"\nAcademic Average Statistics:")
if 'academic_average' in attitude.columns:
    print(f"  Attitude - Min: {attitude['academic_average'].min():.2f}, Max: {attitude['academic_average'].max():.2f}")
if 'academic_average' in research.columns:
    print(f"  Research - Min: {research['academic_average'].min():.2f}, Max: {research['academic_average'].max():.2f}")

# Check GPA values
print(f"\nGPA Statistics:")
if 'gpa' in attitude.columns:
    print(f"  Attitude - Min: {attitude['gpa'].min():.2f}, Max: {attitude['gpa'].max():.2f}")
if 'gpa' in research.columns:
    print(f"  Research - Min: {research['gpa'].min():.2f}, Max: {research['gpa'].max():.2f}")
if 'gpa' in habits.columns:
    print(f"  Habits - Min: {habits['gpa'].min():.2f}, Max: {habits['gpa'].max():.2f}")

# Check department/branch values
print(f"\nDepartment values:")
if 'department' in attitude.columns:
    print(f"  Attitude: {attitude['department'].nunique()} unique values")
if 'department' in research.columns:
    print(f"  Research: {research['department'].unique()[:3]}")
if 'department' in habits.columns:
    print(f"  Habits: {habits['department'].nunique()} unique values")
```

```python
# Section 5: Merge Datasets Based on Common Keys
print("\n5. MERGE DATASETS BASED ON COMMON KEYS")
print("-" * 80)

# Strategy: Since we don't have exact matching keys, we'll:
# 1. First merge Attitude + Research (smaller datasets, more similar)
# 2. Then add Habits data (larger dataset) via concatenation with appropriate column alignment

# Merge Attitude and Research datasets on gender and department (if available)
print(f"\nMerging Attitude + Research datasets...")

# Create a concatenated version that combines all datasets with proper feature alignment
# First, select and align common columns from each dataset

common_features = ['gender', 'academic_average', 'gpa']

# Prepare attitude data
attitude_subset = attitude[[col for col in common_features if col in attitude.columns]].copy()
attitude_subset['source'] = 'attitude'
attitude_subset['total_records'] = len(attitude)

# Prepare research data
research_subset = research[[col for col in common_features if col in research.columns]].copy()
research_subset['source'] = 'research'
research_subset['total_records'] = len(research)

# Prepare habits data (select relevant features)
habits_cols = [col for col in habits.columns 
               if any(key in col for key in ['gender', 'academic', 'gpa', 'exam', 'marks', 'study', 'sleep', 'stress'])]
habits_subset = habits[[col for col in habits_cols[:10] if col in habits.columns]].copy() if habits_cols else habits.copy()
habits_subset['source'] = 'habits'
habits_subset['total_records'] = len(habits)

print(f"  Attitude subset shape: {attitude_subset.shape}")
print(f"  Research subset shape: {research_subset.shape}")
print(f"  Habits subset shape: {habits_subset.shape}")

# Merge using inner join on gender (most reliable common feature)
if 'gender' in attitude.columns and 'gender' in research.columns:
    merged_attitude_research = pd.merge(attitude, research, on='gender', how='outer', suffixes=('_attitude', '_research'))
    print(f"\n✓ Attitude + Research merged: {merged_attitude_research.shape}")
else:
    merged_attitude_research = pd.concat([attitude, research], axis=0, ignore_index=True)
    print(f"\n✓ Attitude + Research concatenated: {merged_attitude_research.shape}")

print(f"  Columns after merge: {len(merged_attitude_research.columns)}")
```

```python
# Section 6: Remove Duplicate and Redundant Columns
print("\n6. REMOVE DUPLICATE AND REDUNDANT COLUMNS")
print("-" * 80)

# Combine all datasets - since they have different structures, concatenate them
print(f"\nCombining all three datasets...")

# Align columns - create a master list of all columns from all datasets
all_cols = set(attitude.columns) | set(research.columns) | set(habits.columns)

# Expand dataframes to include all columns
for col in all_cols:
    if col not in attitude.columns:
        attitude[col] = np.nan
    if col not in research.columns:
        research[col] = np.nan
    if col not in habits.columns:
        habits[col] = np.nan

# Add source identifier
attitude['source_dataset'] = 'attitude'
research['source_dataset'] = 'research'
habits['source_dataset'] = 'habits'

# Concatenate all datasets
merged_df = pd.concat([attitude, research, habits], axis=0, ignore_index=True)
print(f"✓ All datasets concatenated: {merged_df.shape}")

# Remove completely duplicate rows
print(f"\nRemoving duplicates...")
duplicates_before = len(merged_df)
merged_df = merged_df.drop_duplicates(subset=[col for col in merged_df.columns if col != 'source_dataset'], keep='first')
duplicates_removed = duplicates_before - len(merged_df)
print(f"  Duplicates removed: {duplicates_removed}")
print(f"  New shape: {merged_df.shape}")

# Identify and handle redundant columns (columns with same/similar data)
print(f"\nIdentifying redundant columns...")

# Columns with similar names (e.g., gpa, cgpa, avg_gpa) - keep only one version
column_groups = {
    'gpa': ['gpa', 'cgpa'],
    'marks': ['marks', 'marks10th', 'marks12th', 'academic_average', 'college_mark'],
    'gender': ['gender'],
    'department': ['department', 'branch', 'major']
}

cols_to_drop = []
for primary, variants in column_groups.items():
    available = [v for v in variants if v in merged_df.columns]
    if len(available) > 1:
        # Keep the first one, mark others for removal
        cols_to_drop.extend(available[1:])
        print(f"  {primary} group - Keeping '{available[0]}', dropping {available[1:]}")

# Drop redundant columns
if cols_to_drop:
    merged_df = merged_df.drop(columns=cols_to_drop, errors='ignore')
    print(f"\n✓ Redundant columns removed: {cols_to_drop}")

print(f"\nFinal shape after removing duplicates: {merged_df.shape}")
```

```python
# Section 7: Handle Missing Values After Merging
print("\n7. HANDLE MISSING VALUES AFTER MERGING")
print("-" * 80)

# Analyze missing values
print(f"\nMissing values analysis before handling:")
missing_count = merged_df.isnull().sum()
missing_pct = (missing_count / len(merged_df)) * 100
missing_df = pd.DataFrame({
    'column': missing_count.index,
    'missing_count': missing_count.values,
    'missing_pct': missing_pct.values
})
missing_df = missing_df[missing_df['missing_count'] > 0].sort_values('missing_pct', ascending=False)

if len(missing_df) > 0:
    print(missing_df.to_string(index=False))
else:
    print("No missing values detected")

# Handle missing values
print(f"\n\nHandling missing values...")

# For numeric columns: fill with median
numeric_cols = merged_df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if merged_df[col].isnull().sum() > 0:
        median_val = merged_df[col].median()
        merged_df[col].fillna(median_val, inplace=True)
        print(f"  {col}: filled with median {median_val:.2f}")

# For categorical columns: fill with mode
categorical_cols = merged_df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    if merged_df[col].isnull().sum() > 0:
        mode_val = merged_df[col].mode()[0] if not merged_df[col].mode().empty else 'Unknown'
        merged_df[col].fillna(mode_val, inplace=True)
        print(f"  {col}: filled with mode '{mode_val}'")

# Drop columns with >50% missing values (if any remain)
high_missing_cols = missing_df[missing_df['missing_pct'] > 50]['column'].tolist()
if high_missing_cols:
    merged_df = merged_df.drop(columns=high_missing_cols)
    print(f"\n✓ Columns with >50% missing values dropped: {high_missing_cols}")

print(f"\nMissing values after handling: {merged_df.isnull().sum().sum()}")
print(f"Final shape: {merged_df.shape}")
```

```python
# Section 8: Create Final ML Dataset
print("\n8. CREATE FINAL ML DATASET")
print("-" * 80)

# Ensure all data types are appropriate for ML
print(f"\nPreparing final ML dataset...")

# Convert categorical columns to numeric if needed (for ML models)
categorical_cols = merged_df.select_dtypes(include=['object', 'category']).columns
print(f"Categorical columns: {categorical_cols.tolist()}")

# Keep categorical columns as is (can be encoded later for specific ML models)
# For now, just ensure numeric columns are properly typed

for col in merged_df.columns:
    if merged_df[col].dtype == 'object':
        # Try to convert to numeric if possible
        try:
            merged_df[col] = pd.to_numeric(merged_df[col], errors='ignore')
        except:
            pass

# Remove source_dataset column (was only for tracking)
if 'source_dataset' in merged_df.columns:
    merged_df = merged_df.drop(columns=['source_dataset'])

print(f"\nFinal ML Dataset Summary:")
print(f"  Shape: {merged_df.shape}")
print(f"  Columns: {len(merged_df.columns)}")
print(f"  Data types:")
print(merged_df.dtypes.value_counts())

print(f"\n\nDataset Info:")
print(f"  Total records: {len(merged_df)}")
print(f"  Total features: {len(merged_df.columns)}")
print(f"  Missing values: {merged_df.isnull().sum().sum()}")

# Display basic statistics
print(f"\n\nFirst few rows of final dataset:")
print(merged_df.head())
```

```python
# Section 9: Save Final Dataset
print("\n9. SAVE FINAL DATASET")
print("-" * 80)

# Save the final merged and cleaned dataset
output_path = Path(r'c:\vscode\CVprojects\kenexaiHackathon\cleaned datasets\student_success_dataset.csv')
output_path.parent.mkdir(parents=True, exist_ok=True)

merged_df.to_csv(output_path, index=False)
print(f"\n✓ Final dataset saved: {output_path}")

# Display final summary
print(f"\n" + "=" * 80)
print("DATA STITCHING COMPLETE - FINAL SUMMARY")
print("=" * 80)

print(f"\nDatasets merged:")
print(f"  • cleaned_attitude_dataset.csv (235 records)")
print(f"  • research_student_cleaned.csv (221 records)")
print(f"  • cleaned_habits_dataset.csv (80,000 records)")

print(f"\nFinal merged dataset:")
print(f"  File: student_success_dataset.csv")
print(f"  Path: cleaned datasets/")
print(f"  Shape: {merged_df.shape}")
print(f"  Records: {len(merged_df):,}")
print(f"  Features: {len(merged_df.columns)}")

print(f"\nDataset Quality:")
print(f"  ✓ Missing values: {merged_df.isnull().sum().sum()}")
print(f"  ✓ Duplicate rows: {merged_df.duplicated().sum()}")
print(f"  ✓ Data types: {merged_df.dtypes.nunique()} types")

print(f"\nColumns in final dataset:")
for i, col in enumerate(merged_df.columns, 1):
    dtype = merged_df[col].dtype
    print(f"  {i:2d}. {col:40s} ({dtype})")

print(f"\n✓ Dataset ready for machine learning and analysis!")
```
